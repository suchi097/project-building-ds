```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
cols = ['City','City Group','Type']
for i in cols:
    df[i] = le.fit_transform(df[i])
```

```python
x = df.drop(['Open Date','revenue','City'], axis = 1)
y = df['revenue']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

Train test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

1
 model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression(fit_intercept=True,positive=True,copy_X= True)
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
mean_squared_error(y_true= y_test,y_pred=y_pred)
```

```python
r2_score(y_true = y_test,y_pred = y_pred)
```

2 model

```python
import xgboost as xg
xgb_regressor = xg.XGBRegressor(n_estimators=200,max_depth = 3)
xgb_regressor.fit(x_train,y_train)
```

```python
# mean squred error
mean_squared_error(y_test,xgb_regressor.predict(x_test))
```

```python
# r2_score
r2_score(y_true=y_test,y_pred = xgb_regressor.predict(x_test))
```

3 model

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr = GradientBoostingRegressor(learning_rate= 0.01, max_depth =  5, n_estimators = 100, subsample = 0.8)
gbr.fit(x_train,y_train)
```

```python
# mse
gb_pred = gbr.predict(x_test)
mean_squared_error(y_true=y_test,y_pred=gb_pred)
```

```python
r2_score(y_true = y_test,y_pred = gb_pred)
```

4 model

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor(max_depth =5, min_samples_leaf = 4,min_samples_split = 10)
dt.fit(x_train,y_train)
```

```python
dt_pred = dt.predict(x_test)
mean_squared_error(y_test,dt_pred)
```

```python
r2_score(y_true = y_test,y_pred=dt_pred)
```
